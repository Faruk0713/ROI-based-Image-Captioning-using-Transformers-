```python
import json
from sacrebleu import corpus_bleu

file_path = r"C:\Users\fa071\Downloads\Cropped_Image_English_Captions.json"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

references = [entry["english_caption"] for entry in data]
hypotheses = [entry["generated_caption"] for entry in data]

bleu = corpus_bleu(hypotheses, [references])

print(f"SacreBLEU score: {bleu.score:.2f}")
```

```python
import json
from sacrebleu.metrics import BLEU

references = [ [ref] for ref in references ]

references = list(zip(*references))

bleu2 = BLEU(effective_order=True, max_ngram_order=2)
score2 = bleu2.corpus_score(hypotheses, references)

bleu3 = BLEU(effective_order=True, max_ngram_order=3)
score3 = bleu3.corpus_score(hypotheses, references)

bleu4 = BLEU(effective_order=True)
score4 = bleu4.corpus_score(hypotheses, references)

print(f"SacreBLEU (1-2 gram): {score2.score:.2f}")
print(f"SacreBLEU (1-3 gram): {score3.score:.2f}")
print(f"SacreBLEU (1-4 gram): {score4.score:.2f}")

```

```python
import json
from sacrebleu import corpus_bleu

file_path = r"C:\Users\fa071\Downloads\Cropped_Image_Hindi_Captions.json"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

references = [entry["reference_caption"] for entry in data]
hypotheses = [entry["generated_caption"] for entry in data]

# Compute standard BLEU
bleu = corpus_bleu(hypotheses, [references])

print(f"Standard SacreBLEU (1–4 gram): {bleu.score:.2f}")
```

```python
import json
from sacrebleu.metrics import BLEU


references = [ [ref] for ref in references ]  # shape: (num, 1)
references = list(zip(*references))          # shape: (1, num)

bleu2 = BLEU(effective_order=True, max_ngram_order=2)
score2 = bleu2.corpus_score(hypotheses, references)

bleu3 = BLEU(effective_order=True, max_ngram_order=3)
score3 = bleu3.corpus_score(hypotheses, references)

bleu4 = BLEU(effective_order=True)
score4 = bleu4.corpus_score(hypotheses, references)

print(f" SacreBLEU (1-2 gram): {score2.score:.2f}")
print(f" SacreBLEU (1-3 gram): {score3.score:.2f}")
print(f" SacreBLEU (1-4 gram): {score4.score:.2f}")
```
